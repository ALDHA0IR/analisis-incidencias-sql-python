```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "mysql+mysqlconnector://root:root@localhost:3306/analisis_incidencias"
)
```

```python
query = """
SELECT *
FROM incidencias_operativas;
"""

df = pd.read_sql(query, engine)
```

```python
df.shape
df.head()
```

```python
df.info()
df.columns
```

```python
df['fecha_reporte'] = pd.to_datetime(df['fecha_reporte'], dayfirst=True)
df['fecha_atencion'] = pd.to_datetime(df['fecha_atencion'], errors='coerce')
df['tiempo_atencion_horas'] = pd.to_numeric(
    df['tiempo_atencion_horas'], errors='coerce'
)
```

```python
df_atendidas = df[df['estado'] == 'Atendida'].copy()
```

```python
df_atendidas.shape
df_atendidas['estado'].unique()
```

```python
df_atendidas['tiempo_atencion_horas'].mean()
```

```python
df_atendidas.groupby('prioridad')['tiempo_atencion_horas'].mean()
```

```python
df_atendidas.groupby('area_responsable')['tiempo_atencion_horas'].mean()
```

```python
df_atendidas['mes'] = df_atendidas['fecha_reporte'].dt.month
df_atendidas.groupby('mes')['tiempo_atencion_horas'].mean()
```
